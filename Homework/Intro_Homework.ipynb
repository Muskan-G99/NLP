```
import pandas as pd
import gzip
```

```
df= pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz', lines= True)
#Zipped File
```

```
df.head()
```

```
dff= df 
```

```
def cv(str):
  d= dict()
  words= str.lower().split(" ")
  for word in words:
    if word in d:
      d[word] = d[word] +1
    else:
      d[word] = 1
  return d
```

```
cv(df.reviewText[0])
```

```
dff['reviewText_count']= dff['reviewText'].apply(cv)
```
