```
def preprocess(str1):
  str1= str1.lower()
  word_tokens= tokenizer.tokenize(str1)
  filtered= [w for w in word_tokens if not w in stop_words]
  final= ''
  for word in filtered:
    if word.isalpha():
      final= final + ' ' + word
  final= final.translate(str.maketrans('', '', string.punctuation))
  return final
```

```
str1= ' It has Good functionality, but it is not durable like old HPs used to have @Good #Best'
```

```
preprocess(str1)
```

```
corpus= []
for i in range (twitter.shape[0]):
  prep= preprocess(twitter.TweetText[i])
  corpus.append(prep)
```
